```
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the Big Dataset
# REPLACE 'merged_data.csv' with the actual filename you uploaded!
filename = 'merged_data.csv'

try:
    df = pd.read_csv(filename)
    print("‚úÖ File loaded successfully!")
    print(f"Total Rows: {len(df)}")
    print("Columns:", df.columns)

    # 2. Filter for BEXIMCO
    # We look for common column names for the stock symbol
    stock_col = None
    for col in ['Scrip', 'Stock Name', 'trading_code', 'symbol']:
        if col in df.columns:
            stock_col = col
            break

    if stock_col:
        target_stock = 'BEXIMCO'
        df_stock = df[df[stock_col] == target_stock].copy()

        # 3. Clean Date
        # Trying standard formats
        df_stock['Date'] = pd.to_datetime(df_stock['Date'], errors='coerce')
        df_stock = df_stock.sort_values(by='Date')

        print(f"\nüìä Data for {target_stock}:")
        print(f" - History Length: {len(df_stock)} days")
        print(f" - Start Date: {df_stock['Date'].min()}")
        print(f" - End Date: {df_stock['Date'].max()}")

        if len(df_stock) > 100:
            # 4. Plot to confirm
            plt.figure(figsize=(12, 5))
            plt.plot(df_stock['Date'], df_stock['Close'])
            plt.title(f"{target_stock} Long-Term History")
            plt.show()
            print("\nüöÄ READY FOR MACHINE LEARNING! Proceed to next step.")
        else:
            print("\n‚ö†Ô∏è Still too small. Did you upload the 93MB file?")
    else:
        print("‚ùå Could not find a 'Stock Name' column. Check column names above.")

except FileNotFoundError:
    print(f"‚ùå File not found: '{filename}'. Please upload the file and check the name.")
```

```
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# --- 1. FEATURE ENGINEERING (Re-doing it for the big dataset) ---
# We need to recreate the features because we just loaded a new file
df_stock['SMA_10'] = df_stock['Close'].rolling(window=10).mean()
df_stock['SMA_50'] = df_stock['Close'].rolling(window=50).mean()

# RSI Calculation
def calculate_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

df_stock['RSI'] = calculate_rsi(df_stock['Close'])

# Create Target (1 = Price goes UP tomorrow, 0 = Price goes DOWN)
df_stock['Tomorrow_Close'] = df_stock['Close'].shift(-1)
df_stock['Target'] = (df_stock['Tomorrow_Close'] > df_stock['Close']).astype(int)

# Drop missing values (the first 50 days of SMA calculation + last row)
df_ready = df_stock.dropna()

# --- 2. TRAIN/TEST SPLIT ---
# We use 'shuffle=False' because this is Time Series.
# We cannot train on 2024 data and test on 2020 data (that's cheating!)
features = ['SMA_10', 'SMA_50', 'RSI']
X = df_ready[features]
y = df_ready['Target']

# Split: 80% for Training, 20% for Testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print(f"Training on {len(X_train)} days, Testing on {len(X_test)} days.")

# --- 3. TRAIN THE MODEL ---
model = LogisticRegression()
model.fit(X_train, y_train)

# --- 4. EVALUATE ---
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"\nüéØ Baseline Model Accuracy: {accuracy:.2%}")
print("\nClassification Report:")
print(classification_report(y_test, predictions))

# --- 5. VISUALIZE THE CONFUSION MATRIX ---
# This tells us: Did we predict 'UP' when it actually went 'DOWN'?
import seaborn as sns
cm = confusion_matrix(y_test, predictions)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted Down', 'Predicted Up'], yticklabels=['Actual Down', 'Actual Up'])
plt.title('Confusion Matrix (Logistic Regression)')
plt.show()
```

```
# 1. Install the "Pro" libraries
!pip install xgboost shap

import xgboost as xgb
import shap
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2. Train the XGBoost Model (The Kaggle Winner's Algorithm)
# We use a lower 'learning_rate' to prevent overfitting
model_xgb = xgb.XGBClassifier(
    n_estimators=200,      # Number of "trees" in the forest
    learning_rate=0.05,    # Learn slowly to be more precise
    max_depth=5,           # How complex can the rules be?
    random_state=42
)

print("Training XGBoost...")
model_xgb.fit(X_train, y_train)

# 3. Evaluate
predictions_xgb = model_xgb.predict(X_test)
accuracy_xgb = accuracy_score(y_test, predictions_xgb)

print(f"\nüöÄ XGBoost Model Accuracy: {accuracy_xgb:.2%}")
print("\nClassification Report (Look at Class 1 now!):")
print(classification_report(y_test, predictions_xgb))

# 4. EXPLAINABLE AI (The "Why")
# This generates the SHAP graph which explains *how* the AI made decisions
explainer = shap.Explainer(model_xgb)
shap_values = explainer(X_test)

# Plot the "Summary Plot" - The most famous chart in modern AI papers
plt.title("What drives the BEXIMCO Stock Price?")
shap.summary_plot(shap_values, X_test, show=False)
plt.show()
```

```
# 1. Add "Lagged" Features (The most powerful predictor in Finance)
# "Return_Lag_1" = How much did the price change yesterday?
# "Return_Lag_2" = How much did it change 2 days ago?
df_stock['Return'] = df_stock['Close'].pct_change()
df_stock['Return_Lag_1'] = df_stock['Return'].shift(1)
df_stock['Return_Lag_2'] = df_stock['Return'].shift(2)

# 2. Add "Volatility" (Standard Deviation)
# High volatility often means a crash is coming
df_stock['Volatility_5'] = df_stock['Return'].rolling(window=5).std()

# 3. Clean the new NaNs (caused by shifting)
df_ready_v2 = df_stock.dropna()

# 4. Re-Define X and y
features_v2 = ['SMA_10', 'SMA_50', 'RSI', 'Return_Lag_1', 'Return_Lag_2', 'Volatility_5']
X_v2 = df_ready_v2[features_v2]
y_v2 = df_ready_v2['Target']

# 5. Split again (80/20)
X_train_v2, X_test_v2, y_train_v2, y_test_v2 = train_test_split(X_v2, y_v2, test_size=0.2, shuffle=False)

# 6. Re-Train XGBoost
model_xgb_v2 = xgb.XGBClassifier(n_estimators=300, learning_rate=0.03, max_depth=6, random_state=42)
model_xgb_v2.fit(X_train_v2, y_train_v2)

# 7. Evaluate
predictions_v2 = model_xgb_v2.predict(X_test_v2)
print(f"üöÄ Upgraded Model Accuracy: {accuracy_score(y_test_v2, predictions_v2):.2%}")
print("\nNew Classification Report:")
print(classification_report(y_test_v2, predictions_v2))

# 8. NEW SHAP Graph (Compare this to the old one!)
# This will show you if "Yesterday's Return" is more important than RSI
explainer = shap.Explainer(model_xgb_v2)
shap_values = explainer(X_test_v2)
plt.title("What drives BEXIMCO (Upgraded)?")
shap.summary_plot(shap_values, X_test_v2, show=False)
plt.show()
```

```
import matplotlib.pyplot as plt

# 1. Prepare the Test Data for Simulation
# We grab the specific rows we used for testing (2020-2025)
df_simulation = df_ready_v2.iloc[-len(y_test_v2):].copy()
df_simulation['Prediction'] = predictions_v2  # The 0s and 1s from your model

# 2. Calculate Daily Returns
# "Strategy Return": If AI said Buy (1), we get the stock return. If 0, we get 0% (Cash).
df_simulation['Strategy_Return'] = df_simulation['Return'] * df_simulation['Prediction']

# 3. Calculate "Cumulative Profit" (The Growth of 1 Taka)
df_simulation['Cumulative_Market'] = (1 + df_simulation['Return']).cumprod()
df_simulation['Cumulative_AI'] = (1 + df_simulation['Strategy_Return']).cumprod()

# 4. The Final Scoreboard
market_profit = df_simulation['Cumulative_Market'].iloc[-1] - 1
ai_profit = df_simulation['Cumulative_AI'].iloc[-1] - 1

print(f"üí∞ If you invested 1 Lakh Taka in 2020:")
print(f"   - Buy & Hold Strategy would be worth:  {100000 * (1 + market_profit):,.0f} Tk")
print(f"   - Your AI Bot Strategy would be worth: {100000 * (1 + ai_profit):,.0f} Tk")

# 5. Plot the "Race"
plt.figure(figsize=(12, 6))
plt.plot(df_simulation['Date'], df_simulation['Cumulative_Market'], label='Buy & Hold (Market)', color='gray', alpha=0.5)
plt.plot(df_simulation['Date'], df_simulation['Cumulative_AI'], label='AI Bot (XGBoost)', color='green', linewidth=2)
plt.title(f"Man vs. Machine: BEXIMCO Trading Simulation (2020-2025)")
plt.ylabel("Growth of Investment (1.0 = Break Even)")
plt.legend()
plt.grid(True)
plt.show()
```
